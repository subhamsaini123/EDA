# 1. Univariate Analysis 

```python
df.columns
```

```python
sns.countplot(data=df,x=df['Survived'])
```

```python
# If we want to see acutal values then we use value_counts() . 

df['Survived'].value_counts() 
```

```python
df['Survived'].value_counts().plot(kind = 'bar') 
```

```python
# If we want to check for Pclass then 

sns.countplot(data=df,x=df['Pclass'])
```

```python
df['Pclass'].value_counts() 
```

```python
# If we want to find out percentage then use piechart . 
df['Survived'].value_counts() 
```

```python
df['Survived'].value_counts().plot(kind = 'pie' , autopct = '%.2f') 
```

```python
# If We have Numerical data then we use Histogram
# because it finds the distribution . 
```

```python
plt.hist(df['Age']) 
```

```python
sns.distplot(df['Age']) 
```

```python
sns.distplot(df['Age'] , hist = False) 
```

```python
# 3 . Boxplot 
# for find our outliers 

# 1 . lower fence
# 2 . 25% data
# 3.  IQR(Inter Quartile Range)(75%-25%) 
# 4.  75% data 
# 5.  upper fence 
```

```python
# Barplot (Numerical - Categorical) 
```

```python
sns.barplot(x=df['Pclass'] , y=df['Age'])
```

```python
sns.barplot(x=df['Pclass'] ,y= df['Age'] , hue = df['Sex']) 
```

```python
# Boxplot(Numerical-Categorical) 
```

```python
sns.boxplot(x=df['Sex'] ,y= df['Age'] , hue = df['Survived'])
```

```python
# Distplot(Numerical-Categorical)

sns.distplot(df['Age'])
```

```python
sns.distplot(df[df['Survived']==0]['Age'] , hist = False)
```

```python
# HeatMap(Categorical-Categorical) 
p = pd.crosstab(df['Pclass'] , df['Survived'])
p
```

```python
sns.heatmap(p)
```

```python
((df.groupby('Pclass').mean()['Survived'])*100).plot(kind = 'bar')
```
